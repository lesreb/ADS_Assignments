1. Show the age distribution in the data using a histogram.

```python
age_df = titanic_df['Age']

trace1 = go.Histogram(
    x=age_df,
    opacity=0.75,
    name = "Age",
    marker=dict(color='rgba(99, 62, 156, 0.8)'))

data = [trace1]
layout = go.Layout(barmode='overlay',
                   title='Age Distribution',
                   xaxis=dict(title='Age'),
                   yaxis=dict( title='Count'),
)
fig = go.Figure(data=data, layout=layout)

fig.show()
```

2. Show the age distribution based on gender using a histogram.

```python
male_age_df = titanic_df[titanic_df.Sex == 'male']['Age']
female_age_df = titanic_df[titanic_df.Sex == 'female']['Age']
```

```python

trace1 = go.Histogram(
    x=male_age_df,
    opacity=0.75,
    name = "Male",
    marker=dict(color='rgba(99, 62, 156, 0.8)'))

trace2 = go.Histogram(
    x=female_age_df,
    opacity=0.75,
    name = "Female",
    marker=dict(color='rgba(67, 162, 96, 0.8)'))

data = [trace1,trace2]
layout = go.Layout(barmode='overlay',
                   title='Age Distribution Based on Gender',
                   xaxis=dict(title='Age'),
                   yaxis=dict( title='Count'),
)
fig = go.Figure(data=data, layout=layout)

fig.show()
```

4. Pivot the data setting the column Pclass as the columns and Fare as the values. From the
resulting structure, use a boxplot to show the distribution of the values in its 3 columns.

```python
pivot_df = pd.pivot_table(titanic_df,'Fare',index=['PassengerId'],columns=['Pclass'],aggfunc=np.mean)
```

```python
fig = px.box(pivot_df,title = "Distribution of Fares for each Pclass")
fig.show()
```

5. Graph the value counts of the number of passengers who survived and did not survive
based on gender using a stacked bar graph.


```python
survived_male_df=titanic_df[titanic_df.Survived == 1][titanic_df.Sex == 'male']
survived_female_df=titanic_df[titanic_df.Survived == 1][titanic_df.Sex == 'female']
not_survive_male_df=titanic_df[titanic_df.Survived == 0][titanic_df.Sex == 'male']
not_survive_female_df=titanic_df[titanic_df.Survived == 0][titanic_df.Sex == 'female']
```

```python
trace1 = {
  'x':survived_male_df['Sex'],
  'y': survived_male_df['Sex'].value_counts(),
  'name': 'Did Survive - Male',
   'type': 'bar'
};
trace2 = {
  'x':survived_female_df['Sex'],
  'y': survived_female_df['Sex'].value_counts(),
  'name': 'Did Survive - Female',
  'type': 'bar'
};

trace3 = {
  'x':not_survive_male_df['Sex'],
  'y': not_survive_male_df['Sex'].value_counts(),
  'name': 'Did not Survive - Male',
   'type': 'bar'
};

trace4 = {
  'x':not_survive_female_df['Sex'],
  'y': not_survive_female_df['Sex'].value_counts(),
  'name': 'Did Not Survive - Female',
  'type': 'bar'
};
data = [trace1, trace2,trace3, trace4];
layout = {
  'xaxis': {'title': 'Gender'},
  'barmode': 'stack',
  'title': 'Number of Survivals Based on Gender'
};

fig = go.Figure(dict(data = data, layout = layout))


fig.show()
```
