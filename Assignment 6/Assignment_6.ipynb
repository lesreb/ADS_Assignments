3. The change in population of each African country in the last 3 decades.

```python
africa_df = gapminder_df.query("year >= 1991" and "year < 2021" and "continent == 'Africa'")
africa_df
```

```python
africa_df.dtypes
```

```python
africa_df['year_min_diff'] = africa_df['year'] - 1991
```

```python
africa_df
```

```python
#drop rows with data more than three decades (before year 1991)
africa_df_after_91 = africa_df[africa_df['year_min_diff'] >= 0]
africa_df_after_91
```

```python
#recent year for each African country
africa_df_after_91['recent_year'] = 2007 - africa_df['year'] 
```

```python
#for each country, choose the row with minimum year_min_diff (earliest in last 3 decades) 
#and minimum recent_year (most recent year), minimum diff = 1 --> minimum year = 1992.
#Minimum recent year diff ==0 ----> maximum year == 2007

africa_1992=africa_df_after_91[africa_df_after_91['year_min_diff']==1]
africa_1992.index=np.arange(0,len(africa_1992))
africa_1992
```

```python
africa_2007=africa_df_after_91[africa_df_after_91['recent_year']==0]
africa_2007.index=np.arange(0,len(africa_2007))
africa_2007
```

```python
africa_2007['pop_diff'] = africa_2007['pop'] - africa_1992['pop']
```

```python
#africa_2007.index=np.arange(0,len(africa_2007))
africa_2007
```

```python
africa_2007.columns
```

```python
fig = px.choropleth(africa_2007,locations='iso_alpha',
                   color='pop_diff', hover_name='country',
                   color_continuous_scale=px.colors.sequential.matter)

fig.show()
```
