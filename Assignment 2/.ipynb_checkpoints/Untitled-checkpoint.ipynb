```python
#number of attritions
IBM_Attrition_df['Attrition'].value_counts()
```

```python
#plot: Employees (Left vs Stayed) 
sns.countplot(IBM_Attrition_df['Attrition'])
```

```python
# Age vs Attrition
plt.subplots(figsize=(14,6))
sns.countplot(x='Age',hue='Attrition',data = IBM_Attrition_df)
```

**1. Most employees left IBM at ages of 29 and 31**

**2. Most employees at ages 34 and 35 stayed with IBM**

```python
#value counts for each object data type
for column in IBM_Attrition_df.columns:
    if IBM_Attrition_df[column].dtype == object:
        print(str(column)+ ':'+ str(IBM_Attrition_df[column].unique()))
        print(IBM_Attrition_df[column].value_counts())
        print('-------------------------------------------------------------------------------')
```

```python
#visualised correlation of dataset features
plt.figure(figsize = (16,16))
sns.heatmap(IBM_Attrition_df.corr(), annot = True, fmt = '.0%')
```

**1. Percentage salary hike is influenced by Performance rating**

**2. Majority of employees with more years at the company stayed more years with current manager and in their current role**

**3. Monthly income and job level are largely influenced by number of years with the company**

```python
#Analysis of data distribution through histograms
IBM_Attrition_df.hist(figsize=(20,20))
plt.show()
```

**1. Most Employees are between 30 and 40 years**

**2. Most Employees stay closer to workplace**

**3. There are employees at the middle education level**

**4. Job, environment and relationship satisfaction are high for most of the employees**

```python
#Distance from home and attrition
plt.figure(figsize=(14,6))
sns.countplot(x='DistanceFromHome',hue='Attrition',data = IBM_Attrition_df)
```

```python
#Distance from home by Job role and attrition
sns.relplot(x="DistanceFromHome", y="JobRole", hue="Attrition", data=IBM_Attrition_df)#,col="time")
```

**1. Mostly Sales executives who stay far from the work tend to leave the job**

```python
#Distance from home by Job level and attrition
sns.relplot(x="DistanceFromHome", y="JobLevel", hue="Attrition", data=IBM_Attrition_df)
```

**1. Employees at highest job levels are less likely to leave the job**




```python
#Education and Attrition
plt.figure(figsize=(12,6))
sns.countplot(x='Education',hue='Attrition',data = IBM_Attrition_df)
```

**1. Employees at middle Education level (3) constitutes majority of job leavers**

**2. More highly educated employees stay with the company**

```python
# Education Field and attrition
plt.figure(figsize=(12,6))
sns.countplot(x='EducationField',hue='Attrition',data = IBM_Attrition_df)
```

**1. The number of job leavers for each education field is almost directly proportional to the total number of employees for        the corresponding field.** 

**2. Life Science employees hence contribute a large number of job leavers.**

```python
#Monthly income by education and attrition

sns.relplot(x="Education", y="MonthlyIncome", hue="Attrition", data=IBM_Attrition_df)
```

**1. Monthly income by education is not a clear factor influencing Employees to leave their jobs.**

**2. Job leavers are evenly distributed for this combination**

```python
#Job Satisfaction and Attrition
plt.figure(figsize=(12,6))
sns.countplot(x='JobSatisfaction',hue='Attrition',data = IBM_Attrition_df)
```

**1. Mostly satisfied employees stay with IBM**

```python
#Department and Attrition
plt.figure(figsize=(12,6))
sns.countplot(x='Department',hue='Attrition',data = IBM_Attrition_df)
```

**1. Research and Development Department contibute a large number of leavers due to the large size of department employees**

```python
#Job satisfaction by department and attrition

sns.relplot(x="JobSatisfaction", y="Department", hue="Attrition", data=IBM_Attrition_df)
```

**1. Job satisfaction is not a critical factor for attrition for Research and Development department**

```python
#Years since last promotion by Age and attrition

sns.relplot(x="YearsSinceLastPromotion", y="Age", hue="Attrition", data=IBM_Attrition_df)
```

**1. Younger employees leave the company within a few years of no promotion**

```python
#Years since last promotion by Education and attrition

sns.relplot(x="YearsSinceLastPromotion", y="Education", hue="Attrition", data=IBM_Attrition_df)
```

```python
#Years since last promotion by Education field and attrition

sns.relplot(x="YearsSinceLastPromotion", y="EducationField", hue="Attrition", data=IBM_Attrition_df)
```

**1. Life Sciences and Other employees tend to stay with the company despite longer promotion periods**

**2. Employees from the rest of the education fields leave if not promoted for some years**

```python
#Number of companies worked and Attrition
plt.figure(figsize=(12,6))
sns.countplot(x='NumCompaniesWorked',hue='Attrition',data = IBM_Attrition_df)
```
