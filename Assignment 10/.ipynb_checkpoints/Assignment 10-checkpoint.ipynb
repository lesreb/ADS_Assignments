```python
#make 2 copies of the original dataset
df1=df.copy()
df2=df.copy()
df1
```

```python
#Identifying Categorical features from the dataset

categorical_features=[]
for column in df1.columns:
    if df1[column].dtypes == 'object':
        categorical_features.append(column)
categorical_features
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import LabelEncoder, OneHotEncoder

#--------------------------------------------------------------------------------------------#
#LabelEncoder

#categorical_columns=[]
#labelencoder = LabelEncoder()
#for column in df1.columns:
#    if df1[column].dtypes == 'object':
#        df1[column]=labelencoder.fit_transform(df1[column])
#        categorical_columns.append(column)

#print(f'Categorical Columns: {categorical_columns}')
#X = df1
#X
#---------------------------------------------------------------------------------------------#

#using OneHotEncoder for dummy coding of categorical columns: 'age' and 'gender'

onehotencoder = OneHotEncoder()
ct = ColumnTransformer(transformers=[('encode',onehotencoder,[3,4])],remainder='passthrough')
X=pd.DataFrame(ct.fit_transform(df))
X
```

```python
#compute Correlation Matrix for X
corr = X.corr()
corr
```

```python
#heatmap for the correlation
sns.heatmap(corr)
```

```python
#drop the first 4 columns
cols = [0,1,2,3]
X = X.drop(X.columns[cols],axis=1)
X
```

The first 4 columns have less correlation with other features

```python
#compute Correlation Matrix for X after dropping the first 4 columns
corr = X.corr()
corr
```

```python
sns.heatmap(corr)
```

```python
#the elbow method of identifying the proper number of clusters

from sklearn.cluster import KMeans

wcss = []
for i in range(2, 20):
    km = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
    km.fit(X)
    wcss.append(km.inertia_)
    
plt.plot(range(2, 20), wcss)
plt.title('The Elbow Method', fontsize = 20)
plt.xlabel('No. of Clusters')
plt.ylabel('wcss')
plt.show()
```

```python
# from the elbow plot, k = 3
km = KMeans(n_clusters=3)
```

```python
y_means = km.fit_predict(X)
y_means
```

```python
df['Advert_Type']=y_means
df
```

```python
plt.title('Sales per Advert Type for Each Age Bracket')
sns.barplot(data=df, x="age", y="Approved_Conversion", hue="Advert_Type")
```

Advet Type 1 lead to the highest and consistent amount of sales by customers of all the age brackets.

```python
plt.title('Ad Spending For Each Gender')
sns.barplot(data=df, x="gender", y="Spent", hue="Advert_Type")
```

Ad spending for all Advert Type is evenly distributed amongst both genders so their products are gender neutral

```python
import scipy.cluster.hierarchy as shc

plt.figure(figsize=(10, 7))
plt.title("Company xyz Dendograms")
dend = shc.dendrogram(shc.linkage(X, method='ward'))
```

The number of usable clusters is 3

```python
#using 3 clusters in Agglomerative Clustering

from sklearn.cluster import AgglomerativeClustering

cluster = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='ward')
y_means = cluster.fit_predict(X)
y_means
```

```python
df2['Advert_Type']=y_means
df2
```

```python
plt.title('Sales per Advert Type for Each Age Bracket')
sns.barplot(data=df2, x="age", y="Approved_Conversion", hue="Advert_Type")
```

Advet Type 1 lead to the highest and consistent amount of sales by customers of all the age brackets.

```python
plt.title('Ad Spending For Each Gender')
sns.barplot(data=df2, x="gender", y="Spent", hue="Advert_Type")
```

Ad spending for all Advert Type is evenly distributed amongst both genders so their products are gender neutral
